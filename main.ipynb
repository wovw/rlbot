```python
import rlgym
from stable_baselines3 import PPO
from rlgym_tools.sb3_utils import SB3SingleInstanceEnv

from rlgym.utils.obs_builders import AdvancedObs
from terminal import CustomTerminalCondition
from reward import CustomRewardFunction
from action import DiscreteAction
```

```python
# environment vars

ep_len_minutes = 4.0
game_speed = 100
tick_skip = 8
spawn_opponents = True
random_resets = False
team_size = 1
reward_fn = CustomRewardFunction()
terminal_conditions = [CustomTerminalCondition()]
obs_builder = AdvancedObs()
action_parser = DiscreteAction()
```

```python
### training a new model

rl_env = rlgym.make(
    reward_fn=reward_fn,
    terminal_conditions=terminal_conditions,
    obs_builder=obs_builder,
    action_parser=action_parser,
    use_injector=True,
    self_play=True,
)

env = SB3SingleInstanceEnv(rl_env)
tmp_path = "/tmp/sb3_log_test/"

# model definition
model = PPO("MlpPolicy", env, learning_rate=0.0003, n_steps=2048, batch_size=64, n_epochs=30, gamma=0.995,
            gae_lambda=0.95, clip_range=0.2, clip_range_vf=None, normalize_advantage=True, ent_coef=0.01, vf_coef=0.5,
            max_grad_norm=0.5, use_sde=False, sde_sample_freq=-1, target_kl=None, stats_window_size=100,
            tensorboard_log=tmp_path, policy_kwargs=None, verbose=1, seed=None, device='auto', _init_setup_model=True)

# load existing model
# model = PPO.load("./checkpoints/trained_bot.zip", env=env)

# Train agent
while True:
    model = model.learn(total_timesteps=100_000)
    model.save("./checkpoints/test_bot.zip")
    
env.close()
```

```python
# training existing model

rl_env = rlgym.make(
    reward_fn=reward_fn,
    terminal_conditions=terminal_conditions,
    obs_builder=obs_builder,
    use_injector=True,
    self_play=True,
)

env = SB3SingleInstanceEnv(rl_env)
tmp_path = "/tmp/sb3_log_old/"
# logger = configure(tmp_path, ["stdout", "csv", "tensorboard"])

# load existing model
model = PPO.load("./checkpoints/trained_bot.zip", env=env, tensorboard_log=tmp_path)
# model.set_logger(logger)

# Train agent
while True:
    model.learn(total_timesteps=100_000)
    model.save("./checkpoints/trained_bot.zip")

env.close()
```
